```python
#Load the dataset
df=pd.read_csv("DS_EH_1.csv")
df.head(5)
```

```python
# Calculate the % of our missing data in every column
df.isnull().sum()/df.count()*100
```

```python
# Figure out the type of each column
df.dtypes
```

```python
# Use value_counts() on LaunchSite's column to calculate the # of launches
df['LaunchSite'].value_counts()
```

```python
# Let's study more about the Orbit column
df['Orbit'].value_counts()
```

```python
#Use .value_counts() on the column Outcome to determine the # of landing_outcomes
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes
```

```python
# Study more about the landing outcome types.
for count,i in enumerate(landing_outcomes.keys()):
    print(count,i)
```

```python
# Create a set of the outcome types where the first stage landing was not successful.
unsuccessful_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
unsuccessful_outcomes
```

```python
# landing_class = 0 if bad_outcome
# landing_class = 1 otherwise
# Using lambda function for same outcome
df['Class'] = df['Outcome'].apply(lambda landing_class: 0 if landing_class in unsuccessful_outcomes else 1)
df[['Class']].head(10)
```
